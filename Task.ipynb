```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
# Reading dataset
df = pd.read_excel('task1_2 (1).xlsx', sheet_name='Clean data')
```

```python
# Descover dataset
df = df.iloc[:-1,:9]
df.head(25)
```

```python
# Generating descriptive statistics
df.describe().T
```

```python
# Understanding the Data
fig, axes = plt.subplots(4, 4, figsize=(25, 20))
for i,ax in zip(df.columns,axes.flat):
    sns.histplot(df[i],ax=ax)
    ax.set_title(i)
plt.subplots_adjust(wspace=0.2, hspace=0.5)
plt.show()
```

```python
# Adding Profit Column (Earning - Costing)
df['profit'] = df['EPL'] - df['CPL']
df.head()
```

```python
# Computing pairwise correlation of columns to check the relations
df_corr = df.corr()
df_corr.style.background_gradient(cmap='Greens', axis=0)
```

```python
# all profit among whole product
plt.plot(df['profit'], 'p', alpha=0.6, markersize=10, color='orange')
plt.xlabel('profit')
```

```python
# Printing Max&Min of Profit
print(df['profit'].min())
print(df['profit'].max())
```

```python
# Cost group by Product
cost = df.groupby('product')['CPL'].sum()
cost
```

```python
# Revenue group by Product
revenue = df.groupby('product')['EPL'].sum()
revenue
```

```python
# Profit
profit = revenue - cost
round(profit, 2)
```

```python
# Profit among Products (bar chart)
profit.plot.bar(subplots=False, color=['g', 'b', 'r'])
# fig.savefig("profit.png")
```

```python
# Profit among Products (line chart)
fig, ax = plt.subplots(figsize=(15,7), facecolor='white')
profit.plot(ax=ax)
```

```python
# Cost group by Product & Campaign type
# m=df.groupby(by=['product','campaign type'], as_index=False)['profit'].sum()
cost_by_campaign = df.groupby(['product', 'campaign type'])['CPL'].sum().unstack()
cost_by_campaign
```

```python
# Revenue group by Product & Campaign type
revenue_by_campaign = df.groupby(['product', 'campaign type'])['EPL'].sum().unstack()
revenue_by_campaign
```

```python
# Profit group by Product & Campaign type
profit_by_campaign = revenue_by_campaign - cost_by_campaign
round(profit_by_campaign, 2)
```

```python
# Profit group by Product & Campaign type (bar chart)
ax = profit_by_campaign.plot(kind="bar")
fig = ax.get_figure()
fig.set_size_inches(7, 6)
# fig.savefig("profit_by_campaign.png")
```

```python
# Profit group by Product & Campaign type (line chart)
fig, ax = plt.subplots(figsize=(15,7), facecolor='white')
profit_by_campaign.plot(ax=ax)

## Plotty viz
# status=df.groupby(['product','campaign type'])['profit'].sum().reset_index()
# graph=px.bar(status,x='campaign type',y='profit')
# graph.show()
```

```python
# Net Profit Margin group by Product & Campaign type (revenue - cost / revenue)
net_profit_margin = (revenue_by_campaign - cost_by_campaign) / revenue_by_campaign
round(net_profit_margin, 2)
```

```python
# Net Profit Margin group by Product & Campaign type (bar chart)
ax = net_profit_margin.plot(kind="bar")
fig = ax.get_figure()
fig.set_size_inches(7, 6)
# fig.savefig("profit_by_campaign.png")
```

```python
# Net Profit Margin group by Product & Campaign type (line chart)
ax = net_profit_margin.plot(kind="line")
fig = ax.get_figure()
fig.set_size_inches(7, 6)
# fig.savefig("profit_by_campaign.png")
```
